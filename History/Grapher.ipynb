```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
path1 = "/content/drive/MyDrive/Grapher/TDCNN.xlsx"
```

```python
data1 = pd.read_excel(path1)
data1.head()
```

```python
x = [i for i in range(1,26)]
accuracy = data1['accuracy']
val_accuracy = data1['val_accuracy']
val_loss = data1['val_loss']
loss = data1['loss']
```

```python
markersize = 6
linewidth = 1.5

mpl.style.use('seaborn')
plt_1 = plt.figure(figsize=(5,4),dpi=300)
plt.plot(x, accuracy,color='yellow', marker='o', label='Train Acc', linewidth=linewidth, markersize=markersize)
plt.plot(x, val_accuracy, color='green', marker='D', label='Val Acc', linewidth=linewidth, markersize=markersize)
plt.plot(x, loss, color='red', marker='s', label='Train Loss', linewidth=linewidth, markersize=markersize)
plt.plot(x, val_loss, color='blue', marker='X', label='Val Loss', linewidth=linewidth, markersize=markersize)
plt.axis([0,26,0.38,1.05])
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Accuracy/Loss')
plt.title('T-DCNN Training')

plt.show()
```
